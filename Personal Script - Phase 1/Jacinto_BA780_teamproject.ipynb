```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

- Right graph: When comparing the different company types with the education level, we see that most candidates in training come from private limited companies and from undergraduate degrees. Most of the candidates' companies in the dataset are private limited companies of different sizes.
